```python
def plot_numeric_feature_distribution(df, column_name, num_bins=50, stats_pos='left', ax=None):
    """
    Plots the distribution of a numeric feature in a DataFrame using Matplotlib.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame containing the feature to plot.
    column_name : str
        The name of the column containing the feature to plot.
    num_bins : int (default=50)
        The number of bins to be used in the histogram plot
    stats_pos : str (default='left')
        The horizontal position of the statistical summary. Either 'left' or 'right' arguments accepted.
    ax : matplotlib.AxesSubplot or None, optional (default=None)
        The axis object on which to plot the distribution. If None, a new axis object is created.

    Returns:
    --------
    None
    """

    # Create a new axis object if one is not provided
    if ax is None:
        fig, ax = plt.subplots()

    # Calculate the minimum and maximum values of the column
    col_min = df[column_name].min()
    col_max = df[column_name].max()
    
    # Add some buffer to the maximum value
    buffer = (col_max - col_min) * 0.1
    x_max = col_max + buffer
    x_min = col_min - buffer
    
    # Plot the distribution of the feature
    ax.hist(df[column_name], bins=50, range=(x_min, x_max))

    # Set the x-axis limit to the maximum value plus the buffer
    ax.set_xlim(x_min, x_max)

    # Set the title and axis labels
    ax.set_title(f"Distribution of {column_name}", fontsize=16)
    ax.set_xlabel(column_name, fontsize=14)
    ax.set_ylabel("Count", fontsize=14)

    # add mean and std to plot
    if stats_pos=='left':
        ax.text(0.05, 0.95, f'Mean: {round(df[column_name].mean(), 4)}\nStd: {round(df[column_name].std(), 4)}', horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
    if stats_pos=='right':
        ax.text(0.75, 0.95, f'Mean: {round(df[column_name].mean(), 4)}\nStd: {round(df[column_name].std(), 4)}', horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
```

# Discussion Point

Here I have a decision to make as to the potential outliers in the dataset per feature. One method I can use is to drop all rows that contain a value outside a given percentile for any one of the features (minus time) - say outside the 99th percentile. This would give much more meaningful / insightful distributions of data. However, if the fraudulent activity occurs often outside the chosen percentile range, these transactions will be lost and the model will lose valuable data to be trained on. 

Given that the data is so heavily weighted towards non-fraudulent data, it is imperitive that as many of the fraudulent transactions are kept as possible.

I will use a 99th percentile filter to create a separate df and compare the performance of any model on both the new and original datasets.

I expect this to mean that taking a flat filter across the 99th percentiles may well reduce the accuracy of any trained model because such a filter will further the bias towards non-fraudulent transactions by removing a larger proportion of the fraudulent transactions as compared to the non-fraudulent cases (relative to the overall size of the class).

I now plan to create 3 models: One model based on the overall dataset, one based on the 99th percentile filter and another, **ensemble** model based on the 99th percentile filtering. 

In the last of the three I will create two models, one baased on the data that lies within the 99th percentile filter and another for the data that lies outside. At the point of prediction, the model choice will be made based on whether the row contains any data that lies outside the 99th percentile of any given feature. This should account for the loss of fraudulent transactions when filtering.

These 3 models will all be compared for performance.

```python
def find_outliers(df: pd.DataFrame, columns: list) -> pd.DataFrame:
    """
    Returns a new dataframe that contains the same data as the input dataframe but with the addition of a boolean column expressing whether 
    the row contains any data that lies outside the 99th percentile of the corresponding feature.
    
    Parameters:
    -----------
    df : pd.DataFrame
        The dataframe from which to find outliers.
        
    columns : list
        A list of column names in the dataframe. For each column, the function will select
        any rows where the value in the column is greater than the 99th percentile.
    
    Returns:
    --------
    pd.DataFrame
        A new dataframe with the boolean 99th percentile column added to the input dataframe.
    """
    df_outliers = df.copy()
    for col in columns:
        # Calculate the 99th percentile of the column
        percentile_99 = np.percentile(df[col], 99)
        
        # Create a boolean mask for the rows that are outliers based on the 99th percentile
        mask = df[col] > percentile_99
        
        # Filter the dataframe using the boolean mask
        df_outliers['filter'] = mask
        
    return df_outliers
```
