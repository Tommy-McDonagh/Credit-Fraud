```python
def plot_numeric_feature_distribution(df, column_name, num_bins=50, stats_pos='left', ax=None):
    """
    Plots the distribution of a numeric feature in a DataFrame using Matplotlib.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame containing the feature to plot.
    column_name : str
        The name of the column containing the feature to plot.
    num_bins : int (default=50)
        The number of bins to be used in the histogram plot
    stats_pos : str (default='left')
        The horizontal position of the statistical summary. Either 'left' or 'right' arguments accepted.
    ax : matplotlib.AxesSubplot or None, optional (default=None)
        The axis object on which to plot the distribution. If None, a new axis object is created.

    Returns:
    --------
    None
    """

    # Create a new axis object if one is not provided
    if ax is None:
        fig, ax = plt.subplots()

    # Calculate the minimum and maximum values of the column
    col_min = df[column_name].min()
    col_max = df[column_name].max()
    
    # Add some buffer to the maximum value
    buffer = (col_max - col_min) * 0.1
    x_max = col_max + buffer
    x_min = col_min - buffer
    
    # Plot the distribution of the feature
    ax.hist(df[column_name], bins=50, range=(x_min, x_max))

    # Set the x-axis limit to the maximum value plus the buffer
    ax.set_xlim(x_min, x_max)

    # Set the title and axis labels
    ax.set_title(f"Distribution of {column_name}", fontsize=16)
    ax.set_xlabel(column_name, fontsize=14)
    ax.set_ylabel("Count", fontsize=14)

    # add mean and std to plot
    if stats_pos=='left':
        ax.text(0.05, 0.95, f'Mean: {round(df[column_name].mean(), 4)}\nStd: {round(df[column_name].std(), 4)}', horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
    if stats_pos=='right':
        ax.text(0.75, 0.95, f'Mean: {round(df[column_name].mean(), 4)}\nStd: {round(df[column_name].std(), 4)}', horizontalalignment='left',
        verticalalignment='top', transform=ax.transAxes)
```

# Discussion Point

Here I have a decision to make as to the potential outliers in the dataset per feature. One method I can use is to drop all rows that contain a value outside a given percentile for any one of the features (minus time) - say outside the 99th percentile. This would give much more meaningful / insightful distributions of data. However, if the fraudulent activity occurs often outside the chosen percentile range, these transactions will be lost and the model will lose valuable data to be trained on. 

Given that the data is so heavily weighted towards non-fraudulent data, it is imperitive that as many of the fraudulent transactions are kept as possible.

I will use a 99th percentile filter to create a separate df and compare the performance of any model on both the new and original datasets.

I expect this to mean that taking a flat filter across the 99th percentiles may well reduce the accuracy of any trained model because such a filter will further the bias towards non-fraudulent transactions by removing a larger proportion of the fraudulent transactions as compared to the non-fraudulent cases (relative to the overall size of the class).

I now plan to create 3 models: One model based on the overall dataset, one based on the 99th percentile filter and another, **ensemble** model based on the 99th percentile filtering. 

In the last of the three I will create two models, one based on the data that lies within the 99th percentile filter and another for the data that lies outside. At the point of prediction, the model choice will be made based on whether the row contains any data that lies outside the 99th percentile of any given feature. This should account for the loss of fraudulent transactions when filtering.

These 3 models will all be compared for performance.

```python
import numpy as np

def find_outliers(df: pd.DataFrame, columns: list) -> pd.DataFrame:
    """
    Returns a new dataframe that contains the same data as the input dataframe but with the addition of a boolean column expressing whether 
    the row contains any data that lies outside the 99th percentile of the corresponding feature.
    
    Parameters:
    -----------
    df : pd.DataFrame
        The dataframe from which to find outliers.
        
    columns : list
        A list of column names in the dataframe. For each column, the function will select
        any rows where the value in the column is greater than the 99th percentile.
    
    Returns:
    --------
    pd.DataFrame
        A new dataframe with the boolean 99th percentile column added to the input dataframe.
    """
    df_outliers = df.copy()
    for col in columns:
        # Calculate the 99th percentile of the column
        percentile_99 = np.percentile(df[col], 99)
        
        # Create a boolean mask for the rows that are outliers based on the 99th percentile
        mask = df[col] > percentile_99
        
        # Filter the dataframe using the boolean mask
        df_outliers['filter'] = mask
        
    return df_outliers
```

```python
import pandas as pd

def get_high_corr_features(df, target, threshold):
    """
    Returns a list of feature names whose correlation with the target variable
    exceeds a given threshold.
    
    Parameters:
    -----------
    df : pd.DataFrame
        The input dataframe containing the features and target variable.
        
    target : str
        The name of the target variable column.
        
    threshold : float
        The correlation threshold above which to select the features.
        
    Returns:
    --------
    list
        A list of feature names whose correlation with the target variable exceeds
        the given threshold.
    """
    # Calculate the correlation matrix
    corr = df.corr()[[target]]
    
    # Get the absolute correlation values and select only the feature columns
    corr = corr.abs().drop([target])
    
    # Filter the features by the correlation threshold
    high_corr_features = corr[corr[target] > threshold].index.tolist()
    
    return high_corr_features
```

```python
from sklearn.preprocessing import StandardScaler

def standardize_numeric_features(df, numeric_features=None, target_column=None):
    """
    Scale numeric features in a pandas DataFrame using sklearn.preprocessing.StandardScaler.

    Parameters:
    -----------
    df : pandas.DataFrame
        The input DataFrame to be scaled.
        
    numeric_features : list, optional
        A list of column names to be scaled. If None, all numeric columns are selected.

    target_column : str, optional
        The name of the target column to exclude from scaling.

    Returns:
    --------
    pandas.DataFrame
        The scaled DataFrame.
    """
    # Make a copy of the input DataFrame
    df_copy = df.copy()

    # Get list of numeric features
    if numeric_features is None:
        numeric_features = df_copy.select_dtypes(include=[np.number]).columns.tolist()

    # Remove target column from numeric features
    if target_column in numeric_features:
        numeric_features.remove(target_column)

    # Create a StandardScaler object
    scaler = StandardScaler()

    # Normalize each numeric feature in the DataFrame copy
    for feature in numeric_features:
        df_copy[feature] = scaler.fit_transform(df_copy[[feature]])

    return df_copy
```

## Train / Test splits & Downsampling

Due to the very significant bias in the data towards the non-fraudulent class, I will downsample such that there is an equal representation of both classes in the training dataset.


```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import resample

def train_test_downsample(df, test_size=0.2, random_state=None):
    """
    Split the input dataframe into train and test sets, subsample the train set to balance the class distribution, 
    and return the resulting train and test sets.

    Parameters:
    -----------
    df: pandas.DataFrame
        The input dataframe to be split into train and test sets.

    test_size: float, default=0.2
        The proportion of the dataset to include in the test split.

    random_state: int or None, default=None
        The seed used by the random number generator.

    Returns:
    --------
    X_train_downsampled: pandas.DataFrame
        The feature data for the training set.

    X_test: pandas.DataFrame
        The feature data for the test set.

    y_train_downsampled: pandas.Series
        The target variable for the training set.

    y_test: pandas.Series
        The target variable for the test set.
    """
    # Split the original dataset into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=['Class']), df['Class'], test_size=test_size, random_state=random_state)
    
    # Combine the X_train and y_train sets for resampling
    train_df = pd.concat([X_train, y_train], axis=1)
    
    # Separate majority and minority classes
    majority_class = train_df[train_df.Class==0]
    minority_class = train_df[train_df.Class==1]
    
    # Downsample the majority class to balance the dataset
    majority_downsampled = resample(majority_class,
                                    replace=False, # sample without replacement
                                    n_samples=len(minority_class), # match minority n
                                    random_state=random_state) # reproducible results
    
    # Combine minority class and downsampled majority class
    df_downsampled = pd.concat([minority_class, majority_downsampled])
    
    # Split the downsampled dataset into X_train_downsampled and y_train_downsampled
    X_train_downsampled = df_downsampled.drop(columns=['Class'])
    y_train_downsampled = df_downsampled['Class']
    
    return X_train_downsampled, X_test, y_train_downsampled, y_test
```

```python
rng = np.random.RandomState(123)

# for the main df set
X_train_df, X_test_df, y_train_df, y_test_df = train_test_downsample(df, test_size=0.2, random_state=rng)

# for the 99% df set
X_train_df_99, X_test_df_99, y_train_df_99, y_test_df_99 = train_test_downsample(df_99, test_size=0.2, random_state=rng)


# for the 1% df set
X_train_df_1, X_test_df_1, y_train_df_1, y_test_df_1 = train_test_downsample(df_1, test_size=0.2, random_state=rng)
```
