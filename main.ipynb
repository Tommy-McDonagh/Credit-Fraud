# Target (Fraud) Distribution

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Class', data=df)
_= plt.title('Class Distributions \n (0: No Fraud || 1: Fraud)', fontsize=14)
```

```python
# Class distribution
df['Class'].value_counts(normalize=True)
```

```python
def plot_numeric_feature_distribution(df, column_name, ax=None):
    """
    Plots the distribution of a numeric feature in a DataFrame using Matplotlib.

    Parameters:
    -----------
    df : pandas.DataFrame
        The DataFrame containing the feature to plot.
    column_name : str
        The name of the column containing the feature to plot.
    ax : matplotlib.AxesSubplot or None, optional (default=None)
        The axis object on which to plot the distribution. If None, a new axis object is created.

    Returns:
    --------
    None
    """

    # Create a new axis object if one is not provided
    if ax is None:
        fig, ax = plt.subplots()

    # Plot the distribution of the feature
    ax.hist(df[column_name], bins=50)

    # Set the title and axis labels
    ax.set_title(f"Distribution of {column_name}", fontsize=16)
    ax.set_xlabel(column_name, fontsize=14)
    ax.set_ylabel("Count", fontsize=14)
```

# Time Distribution

```python
plot_numeric_feature_distribution(df, 'Time')
```

# Distributions for the V1-28 features

```python
fig, axes = plt.subplots(nrows=7, ncols=4, figsize=(4*4, 7*4))

# select features by iloc indexing
features_to_plot = df.iloc[:, 1:29]

for idx, col in enumerate(features_to_plot):
    ax = axes.flatten()[idx]
    plot_numeric_feature_distribution(df, col, ax=ax)
    plt.subplots_adjust(hspace=0.4, wspace=0.5)
```
